isin() 함수 연습

```python
import pandas as pd

# Creating a DataFrame with temperature and wind speed data
data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'],
        'Temperature (Celsius)': [20, 18, 25, 22, 19],
        'Wind Speed (km/h)': [15, 12, 20, 18, 16]}

weather_df = pd.DataFrame(data)

print(weather_df)
```

```python
# .isin() 함수로 Date컬럼에서 특정 날짜의 rows 추출

selected_dates = ['2023-01-02', '2023-01-04', '2023-01-05']
selected_weather = weather_df[weather_df['Date'].isin(selected_dates)]

print(selected_weather)
```

```python
#연습

# .isin() 함수로 Date컬럼에서 '2023-01-01' 자료만 추출
selected_dates = ['2023-01-01']
selected_weather = weather_df[weather_df['Date'].isin(selected_dates)]

print(selected_weather)
```

```python
#연습

# .isin() 함수로 Date컬럼에서 '2023-01-01', '2023-01-02' 자료만 추출
selected_dates = ['2023-01-01', '2023-01-02']
selected_weather = weather_df[weather_df['Date'].isin(selected_dates)]

print(selected_weather)
```

.fillna() 함수 연습

```python
import pandas as pd
import numpy as np

# Creating a DataFrame with temperature and humidity data, including NaN values
data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'],
        'Temperature (Celsius)': [20, np.nan, 25, 22, 19],
        'Humidity (%)': [50, 60, np.nan, 55, np.nan]}

weather_df = pd.DataFrame(data)

# Using .fillna() to fill NaN values with specific values
# inplace = False, 원 데이터프레임을 바꾸지 않음.
filled_weather_df = weather_df.fillna({'Temperature (Celsius)': 0, 'Humidity (%)': 70}, inplace=False)

print("Original DataFrame:")
print(weather_df)
print("\nDataFrame after filling NaN values:")
print(filled_weather_df)
```

```python
#연습

#isnull 활용, weather_df의 결측데이터 확인
weather_df.isnull()
```

```python
#연습

#isnull.sum 활용, weather_df의 컬럼별 결측데이터 총합 확인
weather_df.isnull().sum()
```

```python
#연습

# 위 데이터프레임 weather_df 의 결측데이터를 {'Temperature (Celsius)': 5, 'Humidity (%)': 50}로 대체.
# 원본데이터 변화 없음
filled_weather_df = weather_df.fillna({'Temperature (Celsius)': 5, 'Humidity (%)': 50}, inplace=False)

print("Original DataFrame:")
print(weather_df)
print("\nDataFrame after filling NaN values:")
print(filled_weather_df)
```

```python
import pandas as pd
import numpy as np

# Creating a DataFrame with temperature and humidity data, including NaN values
data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'],
        'Temperature (Celsius)': [20, np.nan, 25, 22, 19],
        'Humidity (%)': [50, 60, np.nan, 55, np.nan]}

weather_df = pd.DataFrame(data)

# Using .fillna() to fill NaN values with specific values
# inplace = False, 원 데이터프레임을 바꿈.

filled_weather_df = weather_df.fillna({'Temperature (Celsius)': 0, 'Humidity (%)': 70}, inplace=True)

print("Original DataFrame:")
print(weather_df)
print("\nDataFrame after filling NaN values:")
print(filled_weather_df)
```

```python
#연습

# 위 데이터프레임 weather_df 의 결측데이터를 {'Temperature (Celsius)': 5, 'Humidity (%)': 50}로 대체.
# 원본 데이터 자체를 변화
filled_weather_df = weather_df.fillna({'Temperature (Celsius)': 5, 'Humidity (%)': 50}, inplace=True)

print("Original DataFrame:")
print(weather_df)
print("\nDataFrame after filling NaN values:")
print(filled_weather_df)
```

```python
#추가연습
# seaborn dataset에서 'planets' dataset 로드

import pandas as pd
import seaborn as sns

planets = sns.load_dataset('planets')

print(planets)
```

```python
#추가 연습
# planets 데이터프레임 컬럼별 결측데이터 확인
planets.isnull()
```

```python
planets.isnull().sum()
```

```python
#추가연습
#planets 데이터프레임에서 결측데이터가 가장 많은 컬럼 삭제

planets.drop('mass',axis=1,inplace=True)
```

```python
#추가 연습
# mass 삭제 후 column 이름 확인
#.colums 함수 사용
planets.columns
```

```python
#추가 연습
#'distance' 컬럼 최빈값 출력
planets['distance'].mode()
```

```python
#추가 연습
# planets 데이터프레임 distance 컬럼 결측데이터 최빈값으로 수정
# 제일 마지막 행 5개 출력 .tail 사용

planets['distance'].fillna(planets['distance'].mode()[0],inplace=True)
planets.tail()
```

```python
#barplot으로 시각화

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "planets" dataset
planets = sns.load_dataset('planets')

# Set the style of the plot
sns.set(style="whitegrid")

# Count the number of planets discovered each year
planet_count_by_year = planets['year'].value_counts().reset_index().rename(columns={'year': 'planet_count', 'index': 'discovery_year'})

# Create a bar plot for the number of planets discovered each year
plt.figure(figsize=(12, 6))
sns.barplot(x='discovery_year', y='planet_count', data=planet_count_by_year, palette='viridis')
plt.title('Number of Planets Discovered Each Year')
plt.xlabel('Discovery Year')
plt.ylabel('Number of Planets Discovered')
plt.show()
```
